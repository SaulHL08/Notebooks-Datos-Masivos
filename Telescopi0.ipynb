```python
x = pd.DataFrame()
x.loc[:,0] = df.loc[:,'fArrivalTime']
x.loc[:,1] = df.loc[:,'fPhot']

cls = KMeans(init = 'k-means++', n_clusters=4, n_init=10)
cls.fit(x)
x['predicted_label'] = cls.labels_.astype(int)
x.columns = ['fArrivalTime', 'fPhot', 'label']
x
```

```python
conteo_labels = x['label'].value_counts()
print(conteo_labels)
cero = conteo_labels[0]
uno = conteo_labels[1]
dos = conteo_labels[2]

print(f"Número de ceros: {cero}")
print(f"Número de unos: {uno}")
print(f"Número de doses: {dos}")
```

```python
valores_menor_25 = (x['fPhot'] < 25).sum()
print(f"Número de valores en fPhot menores a 25: {valores_menor_25}")
```

```python
# Contar el número de 0, 1 y 2 en la columna 'label'
num_0 = (x['label'] == 0).sum()
num_1 = (x['label'] == 1).sum()
num_2 = (x['label'] == 2).sum()

# Contar el número de valores en 'fPhot' menores que 25
valores_menor_25 = (x['fPhot'] < 25).sum()

# Calcular la suma de 0, 1 y 2 en 'label' y dividirla entre los valores menores de 25 en 'fPhot'
resultado = (num_0 + num_1 + num_2) / valores_menor_25

print(f"Resultado: {resultado}")
```
