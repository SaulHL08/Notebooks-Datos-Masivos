Árboles de decisión

```python
# Importar las bibliotecas necesarias
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Cargar el conjunto de datos de ejemplo (Iris dataset)
iris = load_iris()
X = iris.data
y = iris.target

# Crear un modelo de árbol de decisión
clf = DecisionTreeClassifier()

# Ajustar el modelo a los datos
clf = clf.fit(X, y)

# Mostrar el árbol de decisión
plt.figure(figsize=(12, 8))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
```

Regresión logística

```python
# Importar las bibliotecas necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression

# Cargar el conjunto de datos de ejemplo (Iris dataset)
iris = load_iris()
X = iris.data[:, :2]  # Tomamos solo las dos primeras características para este ejemplo
y = (iris.target != 0) * 1  # Convertir en un problema de clasificación binaria

# Crear un modelo de regresión logística
clf = LogisticRegression(solver='liblinear')

# Ajustar el modelo a los datos
clf.fit(X, y)

# Crear una malla de puntos para visualizar la frontera de decisión
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Mostrar el resultado de la regresión logística
plt.figure(1, figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.RdBu, alpha=0.6)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.RdBu)
plt.xlabel('Longitud del sépalo')
plt.ylabel('Anchura del sépalo')
plt.title('Regresión Logística en Iris dataset')
plt.show()
```

Regresión lineal

```python
# Importar las bibliotecas necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generar datos de ejemplo
np.random.seed(0)
X = np.random.rand(100, 1)  # Variable independiente
y = 2 * X + 1 + 0.1 * np.random.randn(100, 1)  # Variable dependiente con ruido

# Crear y ajustar el modelo de regresión lineal
model = LinearRegression()
model.fit(X, y)

# Predecir valores con el modelo
X_test = np.linspace(0, 1, 100).reshape(-1, 1)  # Valores de X para hacer predicciones
y_pred = model.predict(X_test)

# Visualizar los datos y la línea de regresión
plt.scatter(X, y, label='Datos de entrenamiento')
plt.plot(X_test, y_pred, color='red', linewidth=3, label='Regresión Lineal')
plt.xlabel('Variable Independiente (X)')
plt.ylabel('Variable Dependiente (y)')
plt.legend()
plt.title('Regresión Lineal en Python')
plt.show()
```

Agrupamiento o Clustering: K-Means

```python
# Importar las bibliotecas necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generar datos de ejemplo
n_samples = 300
n_features = 2
n_clusters = 3
random_state = 42

X, _ = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters, random_state=random_state)

# Crear y ajustar el modelo de K-Means
kmeans = KMeans(n_clusters=n_clusters, random_state=random_state)
kmeans.fit(X)

# Obtener las etiquetas de los clusters y los centroides
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Visualizar los datos y los clusters
plt.figure(figsize=(8, 6))

# Dibujar los puntos de datos por cluster
for i in range(n_clusters):
    plt.scatter(X[labels == i, 0], X[labels == i, 1], label=f'Cluster {i + 1}')

# Dibujar los centroides de los clusters
plt.scatter(centers[:, 0], centers[:, 1], marker='x', s=200, c='red', label='Centroides')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.title('Clustering con K-Means')
plt.legend()
plt.show()
```

Reglas de Asociación: A priori

```python
# Importar las bibliotecas necesarias
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import pandas as pd

# Crear un conjunto de datos de ejemplo
data = {'Transaction': ['T1', 'T2', 'T3', 'T4', 'T5'],
        'Items': ['A, B, D', 'B, C', 'A, B, C', 'A, D', 'B, C, D']}
df = pd.DataFrame(data)

# Convertir la columna "Items" en una lista de elementos
df['Items'] = df['Items'].apply(lambda x: x.split(', '))

# Codificar los elementos como variables binarias (usando hot encoding)
df_encoded = pd.get_dummies(df['Items'].apply(pd.Series).stack()).sum(level=0)

# Aplicar el algoritmo Apriori
frequent_itemsets = apriori(df_encoded, min_support=0.4, use_colnames=True)

# Encontrar reglas de asociación
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)

# Mostrar las reglas de asociación
print("Reglas de Asociación:")
print(rules)
```
